```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
import re
```

```python
csv_data = pd.read_csv('NBA MSc Data - Fight URLs.csv')
```

```python
fight_url_list = csv_data['Fight URL'].tolist()
```

```python
fight_url = []
fighter1_list = []
fighter2_list = []
kd1_list = []
kd2_list = []
sig_str1_list = []
sig_str2_list = []
sig_str_pct1_list = []
sig_str_pct2_list = []
total_str1_list = []
total_str2_list = []
td1_list = []
td2_list = []
td_pct1_list = []
td_pct2_list = []
sub_att1_list = []
sub_att2_list = []
rev1_list = []
rev2_list = []
ctrl1_list = []
ctrl2_list = []
```

```python
def crawler(url_list):
    for i in url_list:
        fight_url.append(i)
        source = requests.get(i).text
        soup = BeautifulSoup(source, 'lxml')
        details = soup.find('table')
        body = details.find('tbody')
        stats_table = body.find_all('td')
        fighter1_list.append(stats_table[0].find_all('a')[0].get_text().strip())
        fighter2_list.append(stats_table[0].find_all('a')[1].get_text().strip())
        kd1_list.append(stats_table[1].find_all('p')[0].get_text().strip())
        kd2_list.append(stats_table[1].find_all('p')[1].get_text().strip())
        sig_str1_list.append(stats_table[2].find_all('p')[0].get_text().strip())
        sig_str2_list.append(stats_table[2].find_all('p')[1].get_text().strip())
        sig_str_pct1_list.append(stats_table[3].find_all('p')[0].get_text().strip())
        sig_str_pct2_list.append(stats_table[3].find_all('p')[1].get_text().strip())
        total_str1_list.append(stats_table[4].find_all('p')[0].get_text().strip())
        total_str2_list.append(stats_table[4].find_all('p')[1].get_text().strip())
        td1_list.append(stats_table[5].find_all('p')[0].get_text().strip())
        td2_list.append(stats_table[5].find_all('p')[1].get_text().strip())
        td_pct1_list.append(stats_table[6].find_all('p')[0].get_text().strip())
        td_pct2_list.append(stats_table[6].find_all('p')[1].get_text().strip())
        sub_att1_list.append(stats_table[7].find_all('p')[0].get_text().strip())
        sub_att2_list.append(stats_table[7].find_all('p')[1].get_text().strip())
        rev1_list.append(stats_table[8].find_all('p')[0].get_text().strip())
        rev2_list.append(stats_table[8].find_all('p')[1].get_text().strip())
        ctrl1_list.append(stats_table[9].find_all('p')[0].get_text().strip())
        ctrl2_list.append(stats_table[9].find_all('p')[1].get_text().strip())

    final_data = pd.DataFrame([fight_url, fighter1_list, fighter2_list, kd1_list, kd2_list, sig_str1_list, sig_str2_list, sig_str_pct1_list, sig_str_pct2_list, 
                              total_str1_list, total_str2_list, td1_list, td2_list, td_pct1_list, td_pct2_list, sub_att1_list,
                              sub_att2_list, rev1_list, rev2_list, ctrl1_list, ctrl2_list])
    final_data = final_data.transpose()
    final_data.columns = ['Fight URL', 'Fighter A', 'Fighter B', 'KD A', 'KD B', 'Sig Str A', 'Sig Str B', 'Sig Str % A', 'Sig Str % B',
                         'Total Str A', 'Total Str B', 'TD A', 'TD B', 'TD % A', 'TD % B', 'Sub Att A', 'Sub Att B', 'Rev A',
                         'Rev B', 'CTRL A', 'CTRL B']
    
    return final_data
```
