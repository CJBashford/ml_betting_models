```python
fighter_url_list = csv_data['Fighter URL'].tolist()
```

```python
def crawler(url_list):
    fighter_url = []
    fighter_name = []
    fighter_record = []
    height = []
    reach = []
    stance = []
    dob = []
    
    for i in url_list:
        fighter_url.append(i)
        source = requests.get(i).text
        soup = BeautifulSoup(source, 'lxml')
        fighter_name.append(soup.find('span', class_='b-content__title-highlight').get_text().strip())
        fighter_record.append(soup.find('span', class_='b-content__title-record').get_text().strip())
        info_box = soup.find_all('li', class_='b-list__box-list-item b-list__box-list-item_type_block')
        height.append(info_box[0].get_text().strip())
        reach.append(info_box[2].get_text().strip())
        stance.append(info_box[3].get_text().strip())
        dob.append(info_box[4].get_text().strip())
    
    final_data = pd.DataFrame([fighter_url, fighter_name, fighter_record, height, reach, stance, dob])
    final_data = final_data.transpose()
    final_data.columns = ['Fighter URL', 'Fighter Name', 'Fighter Record', 'Height', 'Reach', 'Stance', 'DOB']
    
    return final_data
```

```python
data = crawler(fighter_url_list)
```
