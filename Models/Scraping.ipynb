```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
import re
```

```python
source = requests.get('https://www.sherdog.com/fighter/Israel-Adesanya-56374').text
```

```python
source
```

```python
soup = BeautifulSoup(source, 'lxml')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
match = soup.find('span', class_='nickname').get_text()
print(match)
```

```python
fight_match = soup.find('div', class_='module fight_history')
```

```python
fight_match
```

```python
headers = fight_match.find('tr', class_="table_head")
```

```python
headers
```

```python
all_rows = fight_match.find_all('tr')[1:]
```

```python
all_rows
```

```python
last_fight = all_rows[0]
```

```python
last_fight
```

```python
fighter = []
result = []
opponent = []
event_name = []
event_date = []
method = []
rd = []
time = []

for row in all_rows:
    cells = row.find_all('td')
    result.append(cells[0].get_text())
    opponent.append(cells[1].get_text())
    event_name.append(cells[2].find('a').get_text())
    event_date.append(cells[2].find(class_='sub_line').get_text())
    method.append(cells[3].get_text())
    rd.append(cells[4].get_text())
    time.append(cells[5].get_text())
```

```python
final_data = pd.DataFrame([result, opponent, event_name, event_date, method, rd, time])

final_data = final_data.transpose()
final_data.columns = ['Result', 'Opponent', 'Event Name', 'Event Date', 'Method', 'Round', 'Time']
```
